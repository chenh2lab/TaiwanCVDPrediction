```python
#temp_df = df_concat[df_concat['year'] == 2000]
temp_df = df_concat
X_temp = temp_df[features].values
```

```python
reducer = umap.UMAP()
embedding = reducer.fit_transform(X_temp)
embedding.shape
```

```python
import matplotlib.patches as mpatches
# Define color mapping
region_map = {"TNKI":0, "THM":1, "TCN":2, "YCTKP":3, "HT":4}
palette = sns.color_palette("tab10", n_colors=len(region_map))

# Scatter plot
scat = plt.scatter(
    embedding[:, 0],
    embedding[:, 1],
    c=[palette[region_map[x]] for x in temp_df['region']],
    s=1
)

# Create legend handles manually
handles = [
    mpatches.Patch(color=palette[i], label=region)
    for region, i in region_map.items()
]

plt.legend(
    handles=handles,
    title="Region",
    loc='best',           # can use 'upper right', 'lower left', etc.
    frameon=False,
    markerscale=4
)

plt.title("UMAP colored by Region")
plt.xlabel("UMAP 1")
plt.ylabel("UMAP 2")
plt.tight_layout()
plt.show()
```

```python
scat = plt.scatter(embedding[:,0], embedding[:,1], c=temp_df.month, s=1, cmap='Spectral')
cb = plt.colorbar(scat)
cb.set_label('month')
plt.xlabel('UMAP-1'); plt.ylabel('UMAP-2'); plt.title('UMAP colored by month')
plt.tight_layout(); plt.show()
```

```python
scat = plt.scatter(embedding[:,0], embedding[:,1], c=temp_df.rate, s=1, cmap='viridis',vmax=0.2, vmin=0.1)
cb = plt.colorbar(scat)
cb.set_label('cvd_rate')
plt.xlabel('UMAP-1'); plt.ylabel('UMAP-2'); plt.title('UMAP colored by cvd_rate')
plt.tight_layout(); plt.show()
```
