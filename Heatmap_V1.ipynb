```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 6CSV
file_paths = [
    "RandomForest.csv",
    "LightGBM.csv",
    "XGBoost.csv",
    "Ridge.csv",
    "Lasso.csv",
    "SVR.csv",
    "KNN.csv",
    "MLP.csv"
]

# 
model_names = ["RandomForest", "LightGBM", "XGBoost", "Ridge", "Lasso", "SVR", "KNN", "MLP"]

# 
dfs = []
for i, file_path in enumerate(file_paths):
    df = pd.read_csv('./r_squares/'+file_path)
    df['model'] = model_names[i]
    dfs.append(df)

# 
all_data = pd.concat(dfs, ignore_index=True)

# 
pivot_table = all_data.pivot(index='group', columns='model', values='r^2')

# 
pivot_table = pivot_table[model_names]

pivot_table = pivot_table.sort_values(by='RandomForest', ascending=False)

#only include the top-ten
pivot_table = pivot_table[:10]

# 
mask = pivot_table.isna()

plt.figure(figsize=(12, 10))
sns.heatmap(pivot_table, 
            mask=mask,  
            cmap='coolwarm',
            vmin=0,  
            annot=True,
            fmt='.3f',
            cbar_kws={'label': 'R'},
            linewidths=0.5,
            linecolor='gray')

plt.rcParams.update({'font.size': 16})

#plt.title('R Heatmap Across Models and Groups', fontsize=16, pad=20)
plt.xlabel('Models', fontsize=16)
plt.ylabel('Groups', fontsize=16)
plt.tight_layout()

# 
plt.savefig(
    "R2_heatmap.png",   # 
    dpi=300,                    # 300 
    bbox_inches='tight',        # 
    pad_inches=0.1              # 
)

plt.show()
```

```python
# 6CSV
file_paths = [
    "RandomForest_mape.csv",
    "LightGBM_mape.csv",
    "XGBoost_mape.csv",
    "Ridge_mape.csv",
    "Lasso_mape.csv",
    "SVR_mape.csv",
    "KNN_mape.csv",
    "MLP_mape.csv"
]

# 
model_names = ["RandomForest", "LightGBM", "XGBoost", "Ridge", "Lasso", "SVR", "KNN", "MLP"]

# 
dfs = []
for i, file_path in enumerate(file_paths):
    df = pd.read_csv('./mape/'+file_path)
    df['model'] = model_names[i]
    dfs.append(df)

# 
all_data = pd.concat(dfs, ignore_index=True)

# 
pivot_table = all_data.pivot(index='group', columns='model', values='mape')

# 
pivot_table = pivot_table[model_names]

pivot_table = pivot_table.sort_values(by='RandomForest', ascending=True)

#only include the top-ten
pivot_table = pivot_table[:10]

# 
mask = pivot_table.isna()


plt.rcParams.update({'font.size': 16})

plt.figure(figsize=(12, 10))
sns.heatmap(pivot_table, 
            mask=mask,  
            cmap='coolwarm_r',
            vmin=5,  
            annot=True,
            fmt='.3f',
            cbar_kws={'label': 'mape'},
            linewidths=0.5,
            linecolor='gray')

#plt.title('Mape Heatmap Across Models and Groups', fontsize=16, pad=20)
plt.xlabel('Models', fontsize=16)
plt.ylabel('Groups', fontsize=16)
plt.tight_layout()

# 
plt.savefig(
    "mape_heatmap.png",   # 
    dpi=300,                    # 300 
    bbox_inches='tight',        # 
    pad_inches=0.1              # 
)

plt.show()
```
